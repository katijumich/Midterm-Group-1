```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
Standings = pd.read_csv('/content/drive/MyDrive/CSC302/Midterm/constructor_standings.csv')
Constructors = pd.read_csv('/content/drive/MyDrive/CSC302/Midterm/constructors.csv')
```

```python
df = pd.merge(Standings, Constructors, on = 'constructorId', how = 'left')
print(df)
```

```python
df.drop(['constructorStandingsId','raceId','position','positionText','constructorRef','url'],axis =1,inplace = True)
```

```python
df.head()
```

```python
df2 = df.groupby('name')['points'].sum().sort_values(ascending=False)
```

```python
print(df2)
```

```python
df3 = df.loc[df['wins']==1].groupby('name')['wins'].count().sort_values(ascending=False).to_frame().reset_index()
```

```python
print(df3)
```

```python
alltime = df3.head(20)
```

```python
sb.barplot(data=alltime, y = 'name', x = 'wins',
           color = 'cyan', alpha = 1, linewidth = 1, edgecolor = 'black')
plt.title('Top 20 All Time Constructors by Wins')
plt.ylabel('Constructor')
plt.xlabel('Number of F1 Wins')
```

```python
plt.figure(figsize=(20, 10))
fig, ax = plt.subplots()

x = alltime['wins']
y = alltime['name']

ax.scatter(x, y,c ="yellow",
            linewidths = 2,
            marker ="D",
            edgecolor ="red")
```

```python
plt.figure(figsize=(20, 10))
fig, ax = plt.subplots()

x = df['name']
y = df['points']

ax.scatter(x, y,c ="yellow",
            linewidths = 2,
            marker ="^",
            edgecolor ="red")
```
