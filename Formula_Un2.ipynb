```python
import pandas as pd
circuits = pd.read_csv('/content/drive/MyDrive/circuits.csv')
circuits.head()
```

```python
c_df = circuits.groupby(by='country').count().sort_values(by='name', ascending=False).reset_index()
c_df['color_opt'] = "orange"
c_df.loc[c_df['name']==1, 'color_opt'] = "red"
c_df.loc[c_df['name']>=8, 'color_opt'] = 'green'
```

```python
c_df.head()
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(9, 6))
plt.bar(c_df['country'], c_df['name'], color=c_df['color_opt'])
red_patch = mpatches.Patch(color='red', label='1 Track')
orange_patch = mpatches.Patch(color='orange', label='2-7 Tracks')
green_patch = mpatches.Patch(color='green', label='12 Tracks')
black_patch = mpatches.Patch(color='black', label='77 Total Worldwide Tracks')
ax.legend(handles=[red_patch, orange_patch, green_patch, black_patch])
ax = plt.gca()
ax.set(xlabel="F1 Host Country", ylabel="Number of F1 Tracks Raced")
ax.set(title="Number of Formula 1 Race Tracks World Wide 1950-2023")
ax.set_xticklabels(c_df['country'], rotation=90, ha='right')

```
